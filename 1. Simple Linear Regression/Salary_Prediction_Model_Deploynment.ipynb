```python
# Importing the pickle module for deploynment
import pickle
```

```python
# Loading the model to the variable model
model = pickle.load(open("salary_prediction_Saved_model.sav","rb"))
```

```python
# Get input from user to make predictions
experience = int(input("Enter your Work experience in years:"))
# store the predicted value in the result variable
result = model.predict([[experience]])
```
