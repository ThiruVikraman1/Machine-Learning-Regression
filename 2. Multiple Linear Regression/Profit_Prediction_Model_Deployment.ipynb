```python
# Import pickle module for deploynment
import pickle
import pandas as pd
```

```python
Loaded_model = pickle.load(open("Profit_Prediction_Saved_Model.sav",'rb'))
```

```python
# Getting inputs from the user
Rd = float(input("Enter the money spent on R&D: "))
Administration = float(input("Enter money spent on Administraation: "))
Marketing = float(input("Enter the money spent in Marketing: "))
florida = int(input("if your state is florida enter 1 or else 0"))
Newyork = int(input("if your state is New York enter 1 or else 0"))
```

```python
User_data = pd.DataFrame([[Rd,Administration,Marketing,florida,Newyork]],columns=['R&D Spend', 'Administration', 'Marketing Spend', 'State_Florida', 'State_New York'])
result = Loaded_model.predict(User_data)
print("The predicted profit is: ",result)
```
