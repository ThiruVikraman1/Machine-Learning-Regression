```python
# Importing necssary libraries
import pandas as pd
```

```python
# Loading the dataset
dataset = pd.read_csv("50_Startups.csv")
dataset.head()
```

```python
# One-hot encoding to handle categorical data(state)
dataset = pd.get_dummies(dataset,drop_first=True,dtype=int)
dataset.tail()    # displays only last 5 rows in the dataset
```

```python
dataset.columns
```

```python
# Feature selection or inputs
independent = dataset[['R&D Spend', 'Administration', 'Marketing Spend',
       'State_Florida', 'State_New York']]
independent.head()
```

```python
# Target or output 
dependent = dataset[['Profit']]
dependent.head()
```

```python
# Splitting data into training and testing sets
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(independent,dependent,test_size=0.25,random_state=0)
```

```python
# Model creation
from sklearn.svm import SVR
regressor = SVR(kernel='linear',C=0.01)  # c is a hyperparameter used for tuning the model
regressor.fit(X_train,Y_train)
```

```python
regressor.intercept_
```

```python
regressor.n_support_
```

```python
regressor.support_vectors_
```

```python
Y_pred = regressor.predict(X_test)
Y_pred
```

```python
# Evaluation metrics
from sklearn.metrics import r2_score
score = r2_score(Y_test,Y_pred)
score
```

1. First r2_score was 0.90
2. After hyperparameter tuning (c=0.01) r2_score = 0.93, therefore hyperparameter tuning have improved the models performance

```python
# Saving the model
import pickle
filename = "Profit_Prediction_SVM_Saved_Model.sav"
pickle.dump(regressor,open(filename,'wb'))
```
